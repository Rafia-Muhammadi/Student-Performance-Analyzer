```python
import numpy as npf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('Students.csv')
print(students)
```

```python
#Total of Each student
students["Total"] = students["Maths"] + students["Physics"]+students["Chemistry"]
print(students)
```

```python
#Average of each student
students["Average"] =students["Total"]/3;
students
```

```python
#Topper 
students.loc[students["Average"].idxmax()]
```

```python
#Each Subject Topper
students.loc[students["Maths"].idxmax()]
students.loc[students["Physics"].idxmax()]
students.loc[students["Chemistry"].idxmax()]
```

```python
#Average of all students Marks in Each subject
print(students["Maths"].mean())
print(students["Physics"].mean())
print(students["Chemistry"].mean())
```

```python
#Highest Marks in all subjects
print(students["Maths"].max())
print(students["Physics"].max())
print(students["Chemistry"].max())
```

```python
#Lowest Marks in all subjects
print(students["Maths"].min())
print(students["Physics"].min())
print(students["Chemistry"].min())
```

```python
#Weak Students
students[ (students["Maths"] <50) |
          (students["Physics"] < 50) |
          (students["Chemistry"] < 50)
    ]
```

```python
#Pass/Fail
students["Result"] ="Pass"
students.loc [
    (students["Maths"]<50)|
    (students["Physics"] < 50)|
    (students["Chemistry"] < 50),
    "Result"
     ] ="Fail"
print(students)
```

```python
#Grades
def grade(avg):
    if avg >= 90:
        return "A"
    elif avg >=75:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >=50:
        return "D"
    else:
        return "F"
        
students["Grade"]= students["Average"].apply(grade)
print(students)
```

```python
#rank
students["Rank"] = students["Average"].rank(ascending = False).astype(int)
print(students)
```

```python
#top 5 students
students.sort_values(
    "Average",
    ascending = False
).head(5)
```

```python
students.to_csv("students_report.csv",index = False)
```

```python
#Top 10 Students by Average Marks-bar
top10 = students.sort_values( 
    "Average",
    ascending = False
).head(10)
print(top10)

plt.figure(figsize=(10,5))
plt.bar(top10['Name'],top10['Average'])
plt.title("Top 10 Students",fontsize = '20',fontfamily='Times New Roman')
plt.xlabel('Students')
plt.ylabel('Average Marks')
plt.xticks(rotation = 45)
plt.savefig('Top 10 students Graph')
plt.show()
```

#### Marks Distribution (Histogram)

```python
plt.figure(figsize= (8,5))
plt.hist(students["Chemistry"],bins = 10)
plt.title("Chemistry MARKS DISTRIBUTION ")
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.savefig('Chemistry Marks Distribution') 
plt.show()
         
```

#### Pie Chart-PASS VS FAIL

```python
results_counts = students["Result"].value_counts()
plt.figure(figsize=(5,6))
plt.pie(
    results_counts,
    labels= results_counts.index,
    autopct = "%1.1f%%"
)
plt.title("Pass Vs Fail",fontsize="16",fontfamily="Times New Roman")
plt.savefig('Pass Vs Fail- Pie Chart') 
plt.show()
```

#### Rank Vs Average

```python
plt.figure(figsize=(8,5))
plt.scatter(students["Rank"],students["Average"])
plt.title("Rank Vs Average",fontsize="16",fontfamily="Times New Roman")
plt.xlabel("Rank")
plt.ylabel("Average Marks")
plt.savefig('Rank vs Average') 

plt.show()
```

#### Compare Subject Marks

```python
plt.figure(figsize=(7,7))

plt.plot(students["Name"],students["Maths"], label = "Maths")
plt.plot(students["Name"],students["Physics"], label = "Physics")
plt.plot(students["Name"],students["Chemistry"], label = "Chemistry")

plt.title("Students Marks Comparision",fontsize="16",fontfamily="Times New Roman")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.xticks(rotation = 90)
plt.legend()

plt.savefig('Compare Subject Marks') 
plt.show()
```

#### Correlation Heatmap

```python
plt.figure(figsize=(6,4))
corr = students[["Maths","Physics","Chemistry"]].corr()

sns.heatmap(
    corr,
    annot=True
)

plt.title("Students Correlation Heatmap")
plt.show()
```

#### Grade Distribution Count Plot

```python
plt.figure(figsize=(6,4))
sns.countplot(
    data=students,
    x="Grade"
)

plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")

plt.show()
```
